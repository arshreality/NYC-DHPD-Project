```python
msk = np.random.rand(len(df_data_3)) < 0.8
train = df_data_3[msk]
test = df_data_3[~msk]

from sklearn import linear_model
regr = linear_model.LinearRegression()
train_x = np.asanyarray(train[['bldgarea']])
train_y = np.asanyarray(train[['bldgdepth']])
regr.fit (train_x, train_y)
# The coefficients
print ('Coefficients: ', regr.coef_)
print ('Intercept: ',regr.intercept_)
```

```python
plt.scatter(train.bldgarea, train.bldgdepth,  color='blue')
plt.plot(train_x, regr.coef_[0][0]*train_x + regr.intercept_[0], 'or')
plt.xlabel("Building Area")
plt.ylabel("Building Depth")
```
